## EDA - Heatmaps



```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
pok = pd.read_csv('pokemon.csv', header=0, sep=',') 
```

Let's see another visualization technique that can allow up to present three dimensional data, and is quite popular - heatmap. 

This visualization technique usually shows the magnitude of a phenomenon as color in two dimensions. Thus, it can be used to show three variables together. In this type of graph two of the three variables are taken to influence - or explain - the third.

Heatmaps show us values as a function of two dimensions. We will first look at a simple heatmap for the correlation coefficients of a few numerical variables of the pokemons. 

```python
pok2 = pd.DataFrame(pok, columns = ['attack', 'defense', 'speed', 'sp_attack', 'sp_defense'])
sns.heatmap(pok2.corr(), annot=True)
```

Note that the parameter *annot* passed with the value *True* determines that the values themselves will be printed on the heatmap. However, even without these values, the color coding enables us to immediately spot high and low values.

With heatmaps, the magnitudes are encoded using either colors or intensity. We can choose another colormap, instead of the defeaul one. If we want to use intensities, we can create a color palette, and use is as a colormap for the heatmap:

```python
colormap = sns.color_palette("Greens") #creating a colormap of green shades
sns.heatmap(pok2.corr(), annot = False ,cmap=colormap) #using the created colormap
```

Another quite popular colormap, maps larger values to reds, and samller values to blues:

```python
sns.heatmap(pok2.corr(), cmap='coolwarm') 
```

Note the usage of the parameter *linewidth* in the following heatmap, which allows us to determine the width of the gap between adjacent cells in the heatmap:

```python
ax = sns.heatmap(pok2.corr(),annot = True ,linewidth=10)
plt.show()
```

We can also make an heatmap using the function *imshow* of matplotlib:

```python
plt.imshow(pok2.corr(), cmap='hot')
plt.show()
```

### Summary

Heatmaps are a nice way to present values as a function of two other dimensions. The intuitive layout allows us to identify patterns and extreme values.
